```python
from src.lib import read_csv_file, get_std, get_mean, get_median, get_all_statistics, create_visualization
```

```python
raw_data = read_csv_file('./StudentPerformanceFactors.csv')
assert not raw_data.empty
assert raw_data.shape == (6607, 20)
raw_data.head()
```

```python
col_name = 'Attendance'
data_mean = get_mean(raw_data, col_name)
data_median = get_median(raw_data, col_name)
data_std = get_std(raw_data, col_name)
summary_stat = raw_data.describe()
assert summary_stat.loc["mean", col_name] == data_mean
assert summary_stat.loc["50%", col_name] == data_median
assert summary_stat.loc["std", col_name] == data_std

print(f"For the column \"Attendance\", its mean is {data_mean:.2f}, its median is {data_median:.2f}, and its standard deviation is {data_std:.2f} ")
```

```python
stats_data = get_all_statistics(raw_data)
create_visualization(raw_data, stats_data, jupyter=True)
```
